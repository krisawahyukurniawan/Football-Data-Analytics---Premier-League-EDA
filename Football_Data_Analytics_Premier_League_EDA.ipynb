# **Analyze Data**

## **Top 10 Players by Goals per 90 Minutes**

```python
# Get the top 10 players
top_10_players = player_season_stats.sort_values(
    by=[('Per 90 Minutes', 'Gls')], ascending=False
).head(10)

# Extract player names and goals per 90 minutes
player_names = [' '.join(map(str, name)) for name in top_10_players.index]
goals_per_90 = top_10_players[('Per 90 Minutes', 'Gls')]

# Create a colormap
colors = cm.viridis(np.linspace(0, 1, len(player_names)))

# Create the bar plot with different colors
plt.figure(figsize=(12, 6))
bars = plt.bar(player_names, goals_per_90, color=colors)
plt.xlabel("Player Names")
plt.ylabel("Goals per 90 Minutes")
plt.title("Top 10 Players by Goals per 90 Minutes")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

# Add data labels on top of each bar
for bar, value in zip(bars, goals_per_90):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(value, 2), va='bottom', ha='center')

plt.tight_layout()
plt.show()
```

```python
# Create a list to store the extracted data
data = []

# Iterate through the DataFrame's index and relevant columns
for index, row in player_season_stats.iterrows():
    try:
        playername = ' '.join(index[3:])  # Joining other parts of the index as the player name
        # Extracting values from multi-index columns (handle potential KeyError)
        performance_min = row[('Playing Time', 'Min')] if ('Playing Time', 'Min') in row else None
        performance_gls = row[('Performance', 'Gls')] if ('Performance', 'Gls') in row else None
        per90_gls = row[('Per 90 Minutes', 'Gls')] if ('Per 90 Minutes', 'Gls') in row else None

        data.append([playername, performance_min, performance_gls, per90_gls])

    except (IndexError, KeyError) as e:
        print(f"Error processing row with index {index}: {e}")


# Create a new DataFrame from the extracted data
per90_goals = pd.DataFrame(data, columns=['Player Name', 'Minutes', 'Total Goals', 'Goals per 90'])

# Display the new DataFrame
per90_goals
```

```python
# sort new_df by Per 90 Minutes, Gls descending
per90_goals.sort_values(by=['Goals per 90', 'Total Goals'], ascending=False, inplace=True)
per90_goals.head(10)
```

## **Top 10 Players by total xG**

```python
# Create a list to store the extracted data
extracted_data = []

# Iterate through the DataFrame's index and relevant columns
for index, row in player_season_stats.iterrows():
    try:
        player = ' '.join(index[3:])  # Joining other parts of the index as the player name
        performance_gls = row[('Performance', 'Gls')] if ('Performance', 'Gls') in row else None
        expected_xG = row[('Expected', 'xG')] if ('Expected', 'xG') in row else None

        extracted_data.append([player, (performance_gls), (expected_xG)])

    except (IndexError, KeyError) as e:
        print(f"Error processing row with index {index}: {e}")

# Create a new DataFrame from the extracted data
goals_vs_xg = pd.DataFrame(extracted_data, columns=['Player', 'Total Goals', 'xG'])
# Display the new DataFrame
goals_vs_xg.head(10)
```

```python
# Sort the DataFrame by xG in descending order and get the top 10
top_10_xg = goals_vs_xg.sort_values(by='xG', ascending=False).head(10)
top_10_xg['Net Goals - xG'] = top_10_xg['Total Goals'] - top_10_xg['xG']

# Display the top 10 players by xG
top_10_xg
```

```python
# Sort the DataFrame by xG in descending order and get the top 10
top_10_xgot = goals_vs_xg.sort_values(by='xG', ascending=False).head(10)

# Display the top 10 players by xG
top_10_xgot
```

## **Top 10 Goalkeeper by Saves Percentage**

```python
goalkeeping = fbref.read_player_season_stats(stat_type="keeper")
goalkeeping.head()
```

```python
# Create a list to store the extracted data
extracted_data = []

# Iterate through the DataFrame's index and relevant columns
for index, row in goalkeeping.iterrows():
    try:
        player = ' '.join(index[3:])
        playing_time_min = row[('Playing Time', 'Min')]
        performance_ga = row[('Performance', 'GA')]
        performance_ga90 = row[('Performance', 'GA90')]
        performance_sota = row[('Performance', 'SoTA')]
        performance_saves = row[('Performance', 'Saves')]
        performance_save_pct = row[('Performance', 'Save%')]

        extracted_data.append([player, playing_time_min, performance_ga, performance_ga90, performance_sota, performance_saves, performance_save_pct])

    except (IndexError, KeyError) as e:
        print(f"Error processing row with index {index}: {e}")

# Create a new DataFrame from the extracted data
goalkeeping_stats = pd.DataFrame(extracted_data, columns=['Player', 'Playing Time (Min)', 'GA', 'GA90', 'SoTA', 'Saves', 'Save%'])

# Display the new DataFrame
goalkeeping_stats
```

```python
# Sort the DataFrame by 'Save%' in descending order and 'Playing Time (Min)' in ascending order
# Filter for players with at least 1000 minutes played
goalkeeping_stats_sorted = goalkeeping_stats[goalkeeping_stats['Playing Time (Min)'] >= 1000].sort_values(by=['Save%', 'Playing Time (Min)'], ascending=[False, True])

# Display the sorted DataFrame
goalkeeping_stats_sorted.head(10)
```

```python
# Create a list to store the extracted data
extracted_data = []

# Iterate through the DataFrame's index and relevant columns
for index, row in player_season_stats.iterrows():
    try:
        player = ' '.join(index[3:])  # Joining other parts of the index as the player name
        performance_gls = row[('Performance', 'Gls')] if ('Performance', 'Gls') in row else None
        performance_g_pk = row[('Performance', 'G-PK')] if ('Performance', 'G-PK') in row else None

        extracted_data.append([player, performance_gls, performance_g_pk])

    except (IndexError, KeyError) as e:
        print(f"Error processing row with index {index}: {e}")

# Create a new DataFrame from the extracted data
extracted_df = pd.DataFrame(extracted_data, columns=['Player', 'Total Goals', 'Non Penalty Goals'])

# Display the new DataFrame
extracted_df.head(10)
```

## **Non Penalty Goals Topscorers**

```python
# Sort the DataFrame by 'Total Goals' in descending order
extracted_df_sorted = extracted_df.sort_values(by='Total Goals', ascending=False)

# Display the sorted DataFrame
extracted_df_sorted.head(10)
```

```python
# Calculate Penalty goals
extracted_df_sorted['Penalty goals'] = extracted_df_sorted['Total Goals'] - extracted_df_sorted['Non Penalty Goals']

# Display the updated DataFrame
extracted_df_sorted.head(10)
```

```python
# Create the stacked bar plot
plt.figure(figsize=(12, 6))
# Players for the plot (adjust the number as needed)
players_to_plot = extracted_df_sorted.head(10)  # Example: Top 10 players
# Define the width of the bars
bar_width = 0.5
# Create the bars
bars1 = plt.bar(players_to_plot['Player'], players_to_plot['Non Penalty Goals'], label='Non-Penalty Goals', width=bar_width, color='skyblue')
bars2 = plt.bar(players_to_plot['Player'], players_to_plot['Penalty goals'], bottom=players_to_plot['Non Penalty Goals'], label='Penalty Goals', width=bar_width, color='orange')

# Add data labels inside the bars
for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval/2, int(yval), va='center', ha='center', color='black', fontsize=8)

for bar in bars2:
    yval = bar.get_height()
    xval = bar.get_x()
    bottom_val = bar.get_y()
    plt.text(xval + bar.get_width()/2, bottom_val + yval/2, int(yval), va='center', ha='center', color='black', fontsize=8)

# Customize the plot
plt.xlabel('Player')
plt.ylabel('Goals')
plt.title('Top 10 goalscorers in the Premier League in 2024-25 season (until GW18)')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
# Display the plot
plt.show()
```

## **Average Possesion Percentage**

```python
possesion = fbref.read_team_season_stats(stat_type="possession")
possesion.head()
```

```python
# Create a list to store the extracted data
extracted_data = []

# Iterate through the DataFrame's index and relevant columns
for index, row in possesion.iterrows():
    try:
        team = index[2]  # Extract team name from the index
        # Access the 'Poss' column directly and convert to float
        poss = float(row['Poss'])
        extracted_data.append([team, poss])
    except (IndexError, KeyError, ValueError) as e:
        print(f"Error processing row with index {index}: {e}")

# Create a new DataFrame from the extracted data
team_possession = pd.DataFrame(extracted_data, columns=['Team', 'Possession'])

# Display the new DataFrame
team_possession
```

```python
team_possession_sorted = team_possession.sort_values(by='Possession', ascending=False)
team_possession_sorted
```

```python
plt.figure(figsize=(14, 6))

# Create a colormap
colors = cm.viridis(np.linspace(0, 1, len(team_possession_sorted)))

# Create a horizontal bar plot
bars = plt.barh(y=team_possession_sorted['Team'], width=team_possession_sorted['Possession'], color=colors)

# Add data labels to the bars
for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height()/2, f'{width:.2f}%', va='center', ha='left', color='black')

# Customize the plot
plt.xlabel('Possession (%)')
plt.ylabel('Team')
plt.title('Premier League Teams by Possession')
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.tight_layout()
plt.show()
```
